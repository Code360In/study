AlexNet
* ImageNet Classification with Deep Convolutional Neural(2012) 
* CNN 구조에 dropout, pooling, ReLu, GPU 기법이 적용됨
* AlexNet은 [Input layer - Conv1 - MaxPool1 - Norm1 - Conv2 - MaxPool2 - Norm2 - Conv3 - Conv4 - Conv5 - Maxpool3 - FC1- FC2 - Output layer] 으로 구성

VGG
* Very Deep Convolutional Networks For Large Scale Image Recognition(2014)
* convolutional networt의 깊이를 증가시키면서 정확도 양상을 조사

GoogleNet
* Going deeper with convolutions(2014)
* 22층을 가진 deep neural network

ResNet
* Deep Residual Learning for Image Recognition(2015)
* Microsoft는 2015년 CNN 구조도 간단하면서 inception v3 보다도 성능이 좋은 Resnet 발표
* residual repesentation 함수를 학습함으로써 신경망이 152 layer까지 가짐

SENet
* Squeeze-and-Excitation Networks(2017)

NASNet
* Learning Transferable Architectures for Scalabel Image Recognition(2018)
senet

EfficientNet
* Rethinking Model Scaling for Convolutional Neural Networks(2019)
* 강화학습으로 최적의 모델을 찾는 MnasNet의 구조를 사용
