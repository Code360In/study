ax.patches 는 ax가 가르키는 그래프에서, 막대들을 담고있는 리스트.

p.get_bbox().bounds 는 해당 막대그래프의 정보를 가져온다.

ax.annotate(s, xy, *args, **kwargs)s : 문자열, xy : 위치, * ha = horizontal align.

위치 값을 x = (left + width)/2 를 해서 가운데로 설정했다. y도 (bottom + height)/2 와 같은 값으로 설정하면 가운데에 정렬할 수 있다.



```
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]

fig, ax = plt.subplots()

ax.bar(X, Y, color='lightgray', edgecolor='gray')

ax.set_xlabel('Salesperson', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)
ax.set_ylabel('Revenue (1000 $)', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)

for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate(height, (left+width/2, height-4), ha = 'center', color = 'k', fontweight = 'bold')
plt.show()

plt.savefig('practice10')
```

```
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]

fig, ax = plt.subplots()

ax.bar(X, Y, color='lightgray', edgecolor='gray')

ax.set_xlabel('Salesperson', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)
ax.set_ylabel('Revenue (1000 $)', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)
for i, p in enumerate(ax.patches):
    h = p.get_height()
    ax.text(i, h+2, format(int(h), ","), ha="center",fontweight = 'bold')
ax.axhline(45, color ='b',linestyle='--')
ax.axhline(15,color='red',linestyle='--')
ax.axhspan(0, 15, facecolor='red',alpha=0.2)
plt.show()
```

bar graph의 zorder = 2, hline의 zorder = 1 로 설정해주니 막대그래프 뒤로 직선이 지나가게 그려졌습니다.

기본적으로 axes에 그려지는 순서는 patches(다양한 도형 컬렉션), 직선, 글자 순서 인데 zorder value에 따라 다르게 그려집니다. 가장 바깥에 그려질수록 zorder가 높아지면 됩니다. 개인적인 생각으로는 x,y 평면에 z축이 꽂혀서 그 위에 레이어를 쌓아가면서 그린다는 느낌입니다.

Axes.axhline(self, y=0, xmin=0, xmax=1, **kwargs)

Axes.axhspan(self, ymin, ymax, xmin=0, xmax=1, **kwargs)

```
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]

fig, ax = plt.subplots()

ax.bar(X, Y, color='lightgray', edgecolor='gray', zorder=2)

ax.set_xlabel('Salesperson', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)
ax.set_ylabel('Revenue (1000 $)', fontsize = 14, fontweight = 'bold', color = 'gray', labelpad=12)

for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate(height, (left+width/2, height-4), ha = 'center', color = 'k', fontweight = 'bold')

ax.axhline(y = 45, color = 'blue', ls = '--', zorder = 1)
ax.axhline(y = 15, color = 'red', ls = '--', zorder = 1)
ax.axhspan(0,15, facecolor='red', alpha=0.2)

plt.show()

plt.savefig('practice10')
```

```
# 2번 : 범주형 데이터 분포
import matplotlib.pyplot as plt
# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.

import seaborn as sns
# seaborn을 sns라는 이름으로 사용합니다.

tips = sns.load_dataset("tips")
# seaborn에 탑재된 tips 데이터셋을 사용합니다.

# label
xlabel = "Day"
ylabel = "Total Bill ($)"
# X축에 "Total Bill ($)"이라는 레이블을, Y축에 "Count"를 달아줍니다.

# label font
label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}
labelpad = 12
# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 12만큼 띄웁니다.

# ticks
tick_labelsize = "large"
# 눈금의 폰트 크기를 크게("large") 키웁니다.

plt.rc('font')

# plot
fig,  ((ax1,ax2,), (ax3,ax4), (ax5,ax6)) = plt.subplots(nrows=3, ncols=2,sharex=True, figsize =(10,12))

sns.barplot(x=tips['day'],y=tips['total_bill'], ax=ax1)
sns.boxplot(x=tips['day'],y=tips['total_bill'], ax=ax2)
sns.swarmplot(x=tips['day'],y=tips['total_bill'], ax=ax3)
sns.swarmplot(x=tips['day'],y=tips['total_bill'], hue = tips['sex'],ax=ax4, dodge=True)
sns.violinplot(x=tips['day'],y=tips['total_bill'], ax=ax5)
sns.violinplot(x=tips['day'],y=tips['total_bill'], hue = tips['sex'],ax=ax6, split=True)
ax6.set_xlabel(xlabel, fontdict = label_fontdict)
ax6.set_ylabel(ylabel,fontdict = label_fontdict)


plt.savefig('파일 이름')
# 파일 이름으로 그림을 저장합니다.
```

```
# 3번 : 데이터 간 관계
import matplotlib.pyplot as plt
# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.

import seaborn as sns
# seaborn을 sns라는 이름으로 사용합니다.

tips = sns.load_dataset("tips")
# seaborn에 탑재된 tips 데이터셋을 사용합니다.

# label
xlabel = "Total Bill ($)"
ylabel = "Tip ($)"
label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}
labelpad = 12
# X, Y축에 각기 "Total Bill ($)"과 "Tip ($)" 라는 레이블을 달아줍니다.
# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 12만큼 띄웁니다.

# ticks
tick_labelsize = "large"
# 눈금의 폰트 크기를 크게("large") 키웁니다.

# plot
fig, axes = plt.subplots(ncols=2, figsize=(10,4))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'], ax=axes[0])
axes[0].set_xlabel(xlabel, fontdict = label_fontdict,labelpad=12)
axes[0].set_ylabel(ylabel, fontdict = label_fontdict,labelpad=12)
sns.lineplot(x=tips['total_bill'],y=tips['tip'], ax=axes[1])
axes[1].set_xlabel(xlabel, fontdict = label_fontdict,labelpad=12, size= tick_labelsize)
axes[1].set_ylabel(ylabel, fontdict = label_fontdict,labelpad=12, size= tick_labelsize)

plt.savefig('파일 이름')
# 파일 이름으로 그림을 저장합니다.
```

```
import matplotlib.pyplot as plt
# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.

import seaborn as sns
# seaborn을 sns라는 이름으로 사용합니다.

penguins = sns.load_dataset("penguins")
# seaborn에 탑재된 penguins 데이터셋을 사용합니다.
penguins["body_mass_100g"] = penguins["body_mass_g"]/100
# 펭귄 체중 데이터를 100g 단위로 표시합니다.
df = penguins.drop('body_mass_g',axis=1)

# label
labels=["Bill Length (mm)", "Bill Depth (mm)", "Flipper Length (mm)", "Body Mass (100g)"]
label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}
labelpad = 12
# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 12만큼 띄웁니다.

# ticks
tick_labelsize = "large"
# 눈금의 폰트 크기를 크게("large") 키웁니다.

# plot
g = sns.PairGrid(df, hue="species",diag_sharey=False)
g.map_diag(sns.kdeplot,fill=True)
g.map_lower(sns.scatterplot)
g.map_lower(sns.regplot, scatter=False, truncate=False, ci=False)
g.map_upper(sns.kdeplot, alpha=0.3)
g.add_legend()
for i in range(4):
    g.axes[3, i].set_xlabel(labels[i], fontdict=label_fontdict)
    g.axes[i, 0].set_ylabel(labels[i], fontdict=label_fontdict)
g.fig.align_ylabels(g.axes[:,0])

plt.savefig('파일 이름')
# 파일 이름으로 그림을 저장합니다.
```

```
# 6번 : 2차원 데이터
import matplotlib.pyplot as plt
# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.

import seaborn as sns
# seaborn을 sns라는 이름으로 사용합니다.

import pandas as pd
# pandas를 pd라는 이름으로 사용합니다.

df = pd.read_csv('/content/drive/MyDrive/dacon/20_temp.csv')
#df = pd.read_csv("20_temp.csv")
# 네브라스카 주 링컨 시의 2016년 일평균 기온 데이터를 가져옵니다.
# 자신의 환경에 맞게 파일 경로(디렉토리)를 지정해주세요.
temp = df.pivot("date", "month", "meantempm")
temp_t = temp.transpose()
# 데이터프레임을 pivot table 형태로 변형하고 행과 열을 바꿉니다.

# label
label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}
labelpad = 12
# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 12만큼 띄웁니다.

# ticks
tick_labelsize = "large"
# 눈금의 폰트 크기를 크게("large") 키웁니다.

# plot
fig, axes = plt.subplots(nrows=4, figsize=(10,16))
sns.heatmap(temp_t,ax= axes[0])
sns.heatmap(temp_t, linewidths=.5, ax= axes[1])
sns.heatmap(temp_t,center=0, ax= axes[2])
sns.heatmap(temp_t,center=0, cmap='RdBu_r',ax=axes[3])

plt.savefig('파일 이름')
# 파일 이름으로 그림을 저장합니다.
```
