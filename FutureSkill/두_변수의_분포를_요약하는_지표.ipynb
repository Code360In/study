```python
### Python package load ###
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

warnings.simplefilter(action='ignore', category=FutureWarning)

# data set load
url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
penguin_data = pd.read_csv(url)

# data pre-processing
penguin_data = penguin_data[~penguin_data.isnull().any(axis=1)]
penguin_data = penguin_data.drop(["island", "year"], axis=1)

# reset plot
plt.clf()


# theme settings
sns.set_style("darkgrid")

fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize = (8,8))
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm' ,data=penguin_data[penguin_data['species'] == 'Adelie'], ax= ax1)
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm' ,data=penguin_data[penguin_data['species'] == 'Gentoo'], ax= ax2)
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm' ,data=penguin_data[penguin_data['species'] == 'Chinstrap'], ax= ax3)
plt.ylim(13,21)
plt.show()
```

# 4번 : 표본 상관계수 구하기

```python
### Python package load ###
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# data set load
url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
penguin_data = pd.read_csv(url)

# data pre-processing
penguin_data = penguin_data[~penguin_data.isnull().any(axis=1)]
penguin_data = penguin_data.drop(["island", "year"], axis = 1)

corr_mat = penguin_data.groupby('species')[['bill_length_mm','bill_depth_mm']].corr()
fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize = (12,6))
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm' ,data=penguin_data[penguin_data['species'] == 'Adelie'], ax= ax1)
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm' ,data=penguin_data[penguin_data['species'] == 'Gentoo'], ax= ax2)
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm' ,data=penguin_data[penguin_data['species'] == 'Chinstrap'], ax= ax3)
plt.ylim(13,21)
print(corr_mat)
plt.show()
```

# 6번 : 표본 상관계수의 의미 이해 2


```python
### Python package load ###
import numpy as np
import pandas as pd

# data set load
url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"
penguin_data = pd.read_csv(url)

# data pre-processing
penguin_data = penguin_data[~penguin_data.isnull().any(axis=1)]
penguin_data = penguin_data.drop(["island", "year"], axis = 1)

x = penguin_data["bill_length_mm"].to_numpy()
y = penguin_data["bill_depth_mm"].to_numpy()

z_x = (x - np.mean(x)) / np.std(x)
z_y = (y - np.mean(y)) / np.std(y)
plt.rcParams["figure.figsize"] = (8,8)
plt.scatter(z_x,z_y)
plt.title('Visualization of Correlation of coefficient')
plt.xlabel('Zx')
plt.ylabel('Zy')
plt.vlines(0,-3,3)
plt.hlines(0,-3,3)
plt.show()
```
