```python
# 2번 : 원하는 정보 추출하기 (1)
import pandas as pd

df_cafes = df[df['상권업종소분류명']== '커피전문점/카페/다방']
df_cafes.head()

#df_cafes = data_set[data_set["상권업종중분류명"].str.contains('커피점/카페')]
#df_cafes.head()
```

```python
# 3번 : 원하는 정보 추출하기 (2)
df_cafe_seoul = df_cafes[df_cafes['시도명'].str.contains('강원도')]
df_cafe_seoul.head()
```

```python
# 4번 : 원하는 정보 추출하기 (3)
df_starbucks  = df[df.상호명.str.contains('스타벅스')]
df_starbucks_seoul  = df_cafe_seoul[df_cafe_seoul.상호명.str.contains('스타벅스')]
```

```python
# 5번 : 원하는 정보 추출하기 (4)
df_starbucks  = df[df.상호명.str.contains('스타벅스')]
df_starbucks_seoul  = df_cafe_seoul[df_cafe_seoul.상호명.str.contains('스타벅스')]

# 투썸플레이스
df_twosome = df[df.상호명.str.contains('투썸플레이스')]
df_twosome_seoul = df_cafe_seoul[df_cafe_seoul.상호명.str.contains('투썸플레이스')]

# 이디야
df_ediya = df[df.상호명.str.contains('이디야')]
df_ediya_seoul = df_cafe_seoul[df_cafe_seoul.상호명.str.contains('이디야')]

# 빽다방
df_bbaek = df[df.상호명.str.contains('빽다방')]
df_bbaek_seoul = df_cafe_seoul[df_cafe_seoul.상호명.str.contains('빽다방')]

# 커피빈
df_coffeebean = df[df.상호명.str.contains('커피빈')]
df_coffeebean_seoul = df_cafe_seoul[df_cafe_seoul.상호명.str.contains('커피빈')]
```
