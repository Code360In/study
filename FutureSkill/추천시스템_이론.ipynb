# 검색
- 검색은 사용자가 입력한 키워드를 기준으로 그것과 관련성 높은 문서를 찾아 사용자에게 보여주는 형태
- 검색은 텍스트 입력이 필요한 검색 창을 전제로 하므로 눈에 잘 보임
- 검색에서 핵심이 되는 데이터 모델은 입력된 키워드, 그리고 그에 상응하는 문서

# 추천
- 사용자가 좋아할 거라고 예측되는 항목(item)을 컴퓨터가 알아서 예측한 후 제시되는 결과
- 눈에 잘 보이지 않음 
- 사용자가 키워드 따위를 직접 입력하지 않기 때문에 검색보다 훨씬 넓고 깊은 데이터 분석을 필요
- 추천에서 중요한 데이터 모델은 사용자와 항목

공통점 1.

수 많은 정보 중 일부를 필터링하는 것

공통점 2.

상위 n개의 결과 중요

--------------------------------------------------------------------------------------

차이점 1.

- 검색은 input으로 들어온 query와 비슷한 documents를 찾는 것

- 추천은 input으로 들어온 user가 좋아할 items을 찾는 것

차이점 2.

- 검색은 사용자의 요청이 있으면 진행

- 추천은 사용자의 요청 없어도 진행

차이점 3.

- 검색은 사용자의 정보가 없어도 문제 없음

- 추천은 사용자의 정보가 많을수록 유용


공통점

1. 검색과 추천은 대상은 다르지만 유사도 분석을 필요로 하는 매칭의 문제



차이점

1. 검색어는 질의어와 문서를 매칭하고, 추천은 사용자와 아이템을 주로 매칭함

2. 검색은 사용자가 입력한 질의어와 관련있는 정보를 제공하며, 추천은 사용자의 선호도를 바탕으로 품목을 제공하여 의사결정을 도움

3. 검색은 사용자가 입력한 질의어가 들어오면 적합한 문서를 찾는 반응적 성격인 반면 추천은 사용자의 요청이 없어도 사전에 품목을 제공함

# CB - Content-based Filtering 
- 유저가 과거에 좋아했던 것과 컨텐츠가 유사한 아이템 추천
- Collaborative Filtering

# 장점 
- User Independence
- CB Filtering은 한 유저의 과거 행동을 분석해 그 유저만을 위한 Profile을 만든다. CF Filtering이 다른 유저의 행동까지 고려해야 한다는 점과는 대조적이다.
- Transparency
추천 시스템이 어떻게 추천 결과를 냈는지에 대한 설명이 가능하다. 추천된 아이템의 특성 벡터를 보면 해당 아이템과 유저가 과거에 좋아했던 아이템이 얼마나 유사한지 볼 수 있다.
- New Item
새로 등록된 아이템에 대해서도 추천이 가능하다

# 단점
- Limited Content Analysis
- CB Filtering은 아이템 추천에 사용할 특성(Features)의 수가 제한적이라는 한계가 있다.
- Over-Specialization
- CB Filtering은 유저가 좋아했던 아이템과 비슷한 아이템만 추천하기 때문에 유저가 예상치 못한 아이템을 추천해줄 수 없다. 이 문제를 Serendipity Problem이라고 한다.
- New User
- CB Filtering이 User Profile을 완성하기까지 충분히 많은 rating이 있어야 한다. 새로 가입한 유저의 경우 신뢰 할 만한 추천을 받기 힘듦.

# Collaborative Filtering
- Rating을 기반으로 유저와 비슷한 성향을 지닌 집단을 찾고, 그들이 좋아하는 상품 추천
- 어떤 종류의 아이템이든지 잘 동작한다. 특징을 선택할 필요가 없다. (No feature selection, only rating vector)

# 장점
- CF는 상호작용의(rating, 구매빈도..) 데이터만 있으면 적용할 수 있다. 따라서 도메인에 의존되지 않고 사용할 수 있다.
- 어떤 종류의 아이템이든지 잘 동작한다. 특징을 선택할 필요가 없다. (No feature selection, only rating vector)
- 일반적으로 content based보다 좋은 성능을 가진다.  

# 단점 
- 하지만 단점으로는 cold start문제가 있다. 
(Cold Start: 시스템에서 맞출만한 충분한 고객들이 필요)
새로운 item이나 user들에 대한 feedback(상호작용 데이터)이 부족하기 때문이다.
- Sparsity: 사용자 또는 등급 행렬 (user/ratings matrix)는 대부분 비어있다. 동일한 아이템을 등급을 매긴 고객을 찾기가 힘들다.
- First rater: 이전에 등급이 매겨지지 않은 아이템을 추천할 수 없다. (새로운 아이템이나 독특한 아이템들)
- Popularity bias: 인기 있는 아이템만 추천하는 경향이 있다. 독특한 취향을 가진 누군가에게 추천할 수 없다.
- 하이브리드 방법: 둘 이상의 다른 추천을 구현하거나 예측을 결합하는 방법으로, 보통 선형 모델을 사용한다.

1. Content-Based Filtering(CB) : item이나 user를 분석하여 비슷한 아이템 추천.

장점 : 적은 정보만으로도 좋은 추천이 가능하다. 많은 양의 user들의 action을 요구하지 않음. cold start 없음

단점 : 모델링 방식에 따라 정확도가 많이 달라지고, 비슷한 아이템 끼리만 추천하여 추천 범위가 제한됨

2. Collaborative Filtering(CF) : item을 분석하지 않고 user의 평가 내역 이용

장점 : 복잡한 아이템을 어렵게 분석하지 않아도 됨. 성능이 좋음. 사용자의 행동 패턴에 따라 적절한 추천이 가능.

단점 : Cold start 문제가 발생. 즉, 수집된 정보의 양이 많아야 좋은 정보가 나오기 때문에 유저의 평가내역 필요.

# 5번 : Memory-based vs Model-based
아래의 설명을 이해한 뒤, Memory-based 방법과 Model-based 방법의 차이점을 말해주세요.

설명
추천 방식 (Recommendation Approach)은 이전에 살펴본 것처럼 CB와 CF가 있습니다.

추천 기술 (Recommendation Technique)은 Memory-based와 Model-based로 나눌 수 있습니다.

① Memory-based

Memory-based algorithms make rating predictions based on the similarities between users or items in the collections of ratings.

② Model-based

Model-based algorithms use the collection of ratings to learn a model, which is then used to make rating predictions.



# Memory-based 방법
- memory-based은 구현하는데 쉽다는 장점
때문에 Amazon과 같은 상업적인 시스템에 효율적으로 사용
- 공통 항목(common item)을 기본으로 하고 있기 때문에 데이터가 희박하다거나 공통된 항목이 적다고 하면 신뢰도가 떨어진다
- memory-based 방법의 단점들을 극복하기 위한 방법이 model-based 방법
- 1) user based collaborative filtering
    *  사용자의 clusters를 확인하고 특정 사용자의 interactions을 이용하여 같은 cluster에 있는 다른 사용자(유사한 사용자)의 interactions을 예측하는 것
- 2) item-based collaborative filtering
    * 사용자  A가 평가한 아이템의 clusters을 확인하고 이 것을 이용하여 이 아이템과 유사한 다른 아이템에 대한 사용자 A의 interactions을 예측하는 것

# Model-based 방법
- 모델 기반의 협업필터링은 머신러닝이나 데이터 마이닝 기술들을 기반으로 한다. 
- 메모리 기반의 협업필터링에 비해, 심지어 matrix가 sparse할 때도 많은 사용자들에게 많은 아이템들을 추천할 수 있다.

# User-based CF
- 사용자-사용자 간의 유사도를 기준
- 사용자 기반의 협업 필터링에서의 유사도는 두 사용자가 얼마나 유사한 항목(아이템)을 선호했는지를 기준
-  코사인 유사도, 피어슨 유사도가 이용

# Item-based CF
- 아이템-아이템 간의 유사도를 기준 

# 서술 
- 아마존과 넷플릭스를 비롯한 서비스에서는 대부분 Item-based CF을 활용
- 또 적절한 UX를 위해서라면, 유저가 아이템을 평가하는 순간, 다른 아이템을 추천할 수 있어야 하는데, 매 평가시마다 유사도 정보를 업데이트하는 것은 현실적으로 어려움 
- 사용자에 비해 아이템 수가 적기 때문에, 아이템 간의 관계 데이터가 발생할 확률이 높다. 따라서 데이터가 누적될수록 추천의 정확도가 높아질 가능성이 더 높기 때문
