```python
# 여러개의 그림
## subplot
x = np.linspace(-1.4, 1.4, 30)

# subplot(2,2,1)은 subplot(221)로 축약
plt.subplot(2, 2, 1)  # 좌측 상단
plt.plot(x, x)
plt.subplot(2, 2, 2)  #  우측 상단
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  #  좌측 하단
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  #  우측 하단
plt.plot(x, x**4)
plt.show()

## fig,ax
x = np.linspace(-1.4, 1.4, 30)
fig, ax = plt.subplots(2, 2) 
ax[0, 0].plot(x, x)       #  좌측 상단
ax[0, 1].plot(x, x**2)   #  우측 상단
ax[1, 0].plot(x, x**3)   #  좌측 하단
ax[1, 1].plot(x, x**4)   #  우측 하단
plt.show()
```

```python
## 일반
from numpy.random import randn

# Axis를 인자로 전달하여 함수 연산과 시각화를 수행합니다.
def plot_line(axis, slope, intercept, **kargs):
    xmin, xmax = axis.get_xlim()
    axis.plot([xmin, xmax], [xmin*slope+intercept, xmax*slope+intercept], **kargs)

x = randn(1000)
y = 0.5*x + 5 + randn(1000)*2

fig, ax = plt.subplots()

ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-5, 15)
ax.scatter(x, y, alpha=0.2)
ax.plot(1, 0, "ro")
ax.vlines(1, -5, 0, color="red")
ax.hlines(0, -2.5, 1, color="red")
plot_line(axis=ax, slope=0.5, intercept=5, color="magenta")  
ax.grid(True)
plt.show()
```

## histogram

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_weather = pd.read_csv("end-part2_df.csv", index_col="date")
df_2016 = df_weather.filter(like="2016", axis=0)
df_2016.head()
```

```python
# 1월 데이터만 선택
df_2016.reset_index(inplace=True)
df_2016["month"] = df_2016["date"].apply(lambda s: int(s.split("-")[1]))
df_jan = df_2016.query("month==1")

fig, ax = plt.subplots()
hists = ax.hist(df_jan["meantempm"])
plt.show()
```

```python
## 1. Gaussian KDE 기능 불러오기
#  scipy의 gaussian_kde 도움을 받아야 합니다.
#  matplotlib의 mlab에 있는 GaussianKDE를 사용할 수도 있습니다.
from scipy.stats import gaussian_kde

## 2. Gaussian KDE 적용
#  데이터를 gaussian_kde()에 넣어서 kde 함수를 만듭니다.
#  density plot을 그릴 X를 정의한 후,
#  X를 kde 함수에 넣어 density plot을 그립니다.
kde = gaussian_kde(df_jan["meantempm"])
X = np.linspace(-20, 10, 301)
Y = kde(X)

## 3. 밀도 함수를 그립니다.
#  histogram과 겹쳐 그리기 위해 histogram을 밀도로 그려줍니다.
#  plot() 명령으로 밀도 함수를 선으로 그립니다.
fig, ax = plt.subplots()
hists = ax.hist(df_jan["meantempm"], density=True)
ax.plot(X, Y)
plt.show()
```

```python
## 2.5. 대역폭 수정
#  kde.covariance_factor를 조정해주고,
#  kde.covariance를 다시 계산하는 과정을 추가해야 합니다.
kde = gaussian_kde(df_jan["meantempm"])
kde.covariance_factor = lambda : .1
kde._compute_covariance()
X = np.linspace(-20, 10, 301)
Y = kde(X)


fig, ax = plt.subplots()
hists = ax.hist(df_jan["meantempm"], density=True)
ax.plot(X, Y)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=12, figsize=(10,7), sharex=True, sharey=True)

for i, ax in enumerate(axes, 1):
    sns.kdeplot(df_2016.query(f"month=={i}")["meantempm"], 
                shade=True, color="cornflowerblue", alpha=1,
                legend=False, ax=ax)
    ax.set_xlim(-29, 40)                       # x범위 조정
    ax.text(-28, 0.02, calendar.month_name[i]) # 월 이름 적용
    
    # 불필요한 윤곽선 없애기
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    if i != 12:
        ax.spines["bottom"].set_visible(False)
        ax.tick_params(axis="x", length=0)      # x축 눈금 없애기. 
    
    ax.set_yticks([]) # y축 눈금 없애기
    
    ax.grid(axis="x") # x축 눈금을 기준으로 grid 만들기
    ax.axhline(0, color="gray")
```

```python
fig, axes = plt.subplots(nrows=12, figsize=(10,7), sharex=True, sharey=True)
fig.patch.set_alpha(0)

for i, ax in enumerate(axes, 1):
    sns.kdeplot(df_2016.query(f"month=={i}")["meantempm"], 
                shade=True, color="w", facecolor="cornflowerblue", alpha=1,
                legend=False, ax=ax)
    ax.set_xlim(-29, 40)    
    ax.text(-29, 0.01, calendar.month_name[i], fontdict={"fontsize":"large"}) 
    
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    if i != 12:
        ax.spines["bottom"].set_visible(False)
        ax.tick_params(axis="x", length=0)
    else:
        ax.tick_params(axis="x", labelsize="large")
        ax.set_xlabel("Mean Temperature ($^o$C)", fontdict={"fontsize":"large"})
    
    ax.set_yticks([])    
    
    ax.grid(axis="x")
    ax.axhline(0, color="gray")
    ax.patch.set_alpha(0)
    
fig.subplots_adjust(hspace=-0.5)
```
